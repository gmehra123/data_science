### Option 1- Use mix inputs from file (Conceptual example)
* Create a file with the overall mix by various segments in different sheets

```python
def get_data_frames(path):
  """Function to read multiple sheets of an excel workbook into a list of dataframes.
  Each dataframe is an element of the list

  Parameters
  ----------
  path: str
    path to the excel workbook with multiple sheets

  Returns
  --------
  dfs_list: list
    list of dataframes """
  xls = pd.ExcelFile(path)
  sheets = xls.sheet_names
  dfs_list = []
  for sheet in sheets:
    df = pd.read_excel(path, sheet_name = sheet)
    dfs_list.append(df)
  return(dfs_list)
```

```python
# once we execute the function we get a list of data Frames
res = get_data_frames(path)
```

```python
# creat5ing a code string using looping and fstrings
str_i="""res[0]"""
str_sfnl=""
for i in range(len(res)-1):
  str1 = f""".merge(res[{i+1}],how="cross")"""
  str_sfnl+= str1
```

```python
# Creating the final string to evaluate by adding the first data frame
str_fnl = str_i + str_sfnl
```

```python
# Check that the cartesian product includes all the excel sheets. If
str_fnl
```

```python
# Executing the code string to cross join all data frames
r=eval(str_fnl)
r.head()
```

```python

float_dtypes = (r.dtypes[r.dtypes == 'float']).index
rest = 1
print(float_dtypes)
for col in float_dtypes:
  rest = r[col] * rest
```

```python
r.loc[:,'result'] = rest
```

```python
# In this step we enter the yearly gains, if yearly gains are fixed, for eg, 37.5K across all years. If different it can just be a list
gains = input("Enter yearly gains")
try:
  gains = int(gains)
except:
  print("not an integer")
```

```python
# Now multiply by the annual gains number to distribute it over the segments in the required proportion
r['counts']=r['result']*gains
```
